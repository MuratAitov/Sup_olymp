```python
def parse_row(row):
    spec = row[0]
    privelege_type = None
    if row[-2] == 'Зачисление без вступительных испытаний' :
        privelege_type = 'бви'
    elif "ЕГЭ" in row[-2]:
        privelege_type = 'ЕГЭ'
    elif "дополнительн" in row[-2]:
        privelege_type = 'дви'
    olymps = row[3].split(',')
    faculty = row[-1]
    levels = row[4].split(',')
    class_num = row[5]
    ret = []
    for level in levels:
        ret += [(spec,privelege_type,class_num, level, olymp, faculty) for olymp in olymps]
    return ret
```

```python
big_list = []
for num, dfff in enumerate(dfs):
    for index, row in dfff.iterrows():
        big_list+=parse_row(list(row))
```

```python
big_dict = {}
```

```python
def try_make(dct,key):
    if key not in dct.keys():
        dct[key] = {}
```

```python
def try_list(dct, key):
    if key not in dct.keys():
        dct[key] = []
```

```python
for spec,privelege_type,class_num, level, olymp, faculty in big_list:
    try_make(big_dict, faculty)
    try_make(big_dict[faculty], spec)
    try_list(big_dict[faculty][spec], privelege_type)
    big_dict[faculty][spec][privelege_type].append((class_num, level, olymp))
```

```python
import json
with open('olympiads21.json', 'w') as f:
    json.dump(big_dict, f)
big_dict
```
